```python
# importing the module
import json
import pandas as pd
import config      #python file with API credentials for authentication
import matplotlib
import tweepy
import requests
import matplotlib.pyplot as plt
from tweepy import OAuthHandler
import time
import sys

consumer_key=config.api_key
consumer_secret=config.api_key_secret
access_token_key=config.access_token
access_token_secret=config.access_token_secret

bearer_token=config.bearer_token


client = tweepy.Client(bearer_token, consumer_key, 
                        consumer_secret, access_token_key, 
                        access_token_secret, wait_on_rate_limit=True)

auth = tweepy.OAuth1UserHandler( consumer_key,  consumer_secret, access_token_key, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
ulist=[]
 
for u in user_info['user_id']:
    print(u)
    user = api.get_user(user_id=u)
    try:
        url = user.entities['url']['urls'][0]['expanded_url']
    except:
        url = ''
    udata = [user.id,user.name,user.screen_name, user.location,user.description,user.followers_count,user.listed_count,user.statuses_count,url]
    ulist.append(udata)
 
users_info = pd.DataFrame(ulist,columns=['user_id','name','screen_name','location','description','followers_count','listed_count','statuses_count','url'])
```

```python
len(users_info)
```

```python
tweets_data = []


for u in range (0, len(users_info)):
    print(u)
    for tweet in tweepy.Cursor(api.user_timeline, user_id=users_info['user_id'][u], 
                                    tweet_mode="extended",  exclude_replies=True, include_rts=False).items(80):
            tweet = tweet._json 
            tweet_data = {}

            for e in to_extract:
                try:
                    tweet_data[e]=tweet[e]
                except:
                    continue
            for entity in tweet_entities:
                entity_list = []
                for t in tweet['entities'][entity[0]]:
                    entity_list.append(t[entity[1]])
                tweet_data[entity[0]] = entity_list
        
            tweets_data.append(tweet_data)
            tweet_data['user_id'] = users_info['user_id'][u]
            tweet_data['screen_name']=users_info['screen_name'][u]
            tweet_data['description']=users_info['description'][u]
            tweet_data['user_location']=users_info['location'][u]
    time.sleep(1)

# save all data in a dataframe
data = pd.DataFrame(tweets_data)
```
